```python
import pandas as pd
carDF = pd.read_csv('car_financing.csv', encoding='latin-1')
TSDF = carDF.where(carDF['car_type'] == 'Toyota Sienna').dropna()
sienna_total_balance = TSDF['Starting Balance'].sum()
sienna_total_repayment = TSDF['Repayment'].sum()
sienna_total_new_balance = TSDF['New Balance'].sum()
sienna_average_interest_rate = TSDF['interest_rate'].mean()
print(TSDF.to_string())
print(f'Total: Balance {sienna_total_balance}.')
print(f'Total: Repayment {sienna_total_repayment}.')
print(f'Total: New Balance {sienna_total_new_balance}.')
print(f'Average: Interest Rate {sienna_average_interest_rate}.')
```
